```python
import pandas as pd
from textblob import TextBlob
from nltk.sentiment import SentimentIntensityAnalyzer

# Load dataset
df = pd.read_csv("after analysis.csv", encoding="ISO-8859-1")  # Change encoding if needed
df.head()
```

```python
# Ensure 'title' column exists
#if 'title' not in df.columns:
    #raise ValueError("No 'title' column found in dataset")

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to analyze sentiment
def get_sentiment(text):
    if not isinstance(text, str) or text.strip() == "":
        return "Neutral"  # Default for empty or missing titles
    
    # VADER Sentiment Analysis
    vader_score = analyzer.polarity_scores(text)["compound"]
    
    # TextBlob Sentiment Analysis
    blob_score = TextBlob(text).sentiment.polarity

    # Average Score for better accuracy
    final_score = (vader_score + blob_score) / 2

    # Assign detailed sentiment labels
    if final_score > 0.05:
        if "price" in text.lower():
            return "Price Effective (Pos)"
        elif "features" in text.lower():
            return "Good Features (Pos)"
        elif "design" in text.lower() or "quality" in text.lower():
            return "Good Design & Quality (Pos)"
        elif "user friendly" in text.lower():
            return "User Friendly (Pos)"
        else:
            return "Satisfactory & Recommended (Pos)"
    
    elif final_score < -0.05:
        if "design" in text.lower() or "quality" in text.lower():
            return "Poor Design & Quality (Neg)"
        elif "features" in text.lower() and "limited" in text.lower():
            return "Limited Features (Neg)"
        elif "features" in text.lower():
            return "Bad Features (Neg)"
        elif "difficult" in text.lower() or "hard to use" in text.lower():
            return "Difficult to Use (Neg)"
        elif "money" in text.lower() or "expensive" in text.lower():
            return "Not Worth The Money (Neg)"
        else:
            return "Unsatisfactory (Neg)"
    
    else:
        return "Neutral"

# Apply sentiment analysis to the "title" column
df["sentiment"] = df["Title"].apply(get_sentiment)

# Print updated DataFrame
print(df.head())

# Save results if needed
#df.to_csv("title_sentiment_analysis.csv", index=False)
```
