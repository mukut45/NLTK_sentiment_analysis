```python
import pandas as pd
from textblob import TextBlob # this also from the NLP library
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer # this i am calling it from the library of NLP

# Load dataset
df = pd.read_csv("after analysis.csv", encoding="ISO-8859-1")  # Change encoding if needed

# Ensure 'title' column exists
#if 'title' not in df.columns:
    #raise ValueError("No 'title' column found in dataset")

# Initialize VADER sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Function to analyze sentiment
def get_sentiment(text):
    if not isinstance(text, str) or text.strip() == "":
        return "Neutral"  # Default for empty or missing titles
    
    # VADER Sentiment Analysis
    vader_score = analyzer.polarity_scores(text)["compound"]
    
    # TextBlob Sentiment Analysis
    blob_score = TextBlob(text).sentiment.polarity

    # Average Score for better accuracy
    final_score = (vader_score + blob_score) / 2

    # Assign sentiment label in terms of negetive and positive
    if final_score > 0.05:
        return "Positive"
    elif final_score < -0.05:
        return "Negative"
    else:
        return "Neutral"

# Apply sentiment analysis to the "title" column
df["sentiment"] = df["Title"].apply(get_sentiment)

# Print sample results
print(df[["Title", "sentiment"]].head())
```
